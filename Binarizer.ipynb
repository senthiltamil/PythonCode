```python
import pandas as pd
import numpy as np
def binary_labelizer(col_prefix,series1):
    x_index = list(series1.value_counts().index)
    x_eye = np.eye(len(x_index),dtype=int)
    dic = {index:x_eye[x_index.index(x)] for index,x in series1.iteritems()}
    df = pd.DataFrame(dic).T
    df.columns = [col_prefix + '_' + x for x in x_index]
    return df
```

```python
df = pd.read_csv('binarizationinput.csv')
binary_labelizer('color',df.iloc[:,0])
```
